# Polynomial linear regrssion

왜 선형인가?.. x끼리가 아니라 계수끼리 선형조합으로 표현할 수 있는 형태인 것이라.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import data

```python
data = pd.read_csv('D:/Udemy/machine learning AZ/colab_notebook/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv')
data.head()
```

```python
X = data.iloc[:,1 :-1].values
y = data.iloc[:,-1].values
```

```python
X
```

## Linear regression 해보기

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Poly linear 해보기

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) # 2차항
X_poly = poly_reg.fit_transform(X)
X_poly
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

## Visuallising the Linear regression

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Simple Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

# Visuallising the Poly Linear regression

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Poly Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

## upgrade the degree

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg_4 = PolynomialFeatures(degree=4) # 4차항
X_poly_4 = poly_reg_4.fit_transform(X)
X_poly_4
```

```python
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly_4,y)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_4.predict(X_poly_4), color='blue')
plt.title('Poly Linear Regression_degree_4')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

## Predicting a new result with linear model

```python
lin_reg.predict([[6.5]]) # 실제 그래프 보면 6.5는 0.2 정도 나옴
```

## Predicting a new result with Poly

```python
lin_reg_4.predict(poly_reg_4.fit_transform([[6.5]]))
```
